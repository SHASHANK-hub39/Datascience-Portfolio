```python
data = pd.read_csv('/content/drive/My Drive/Googlecolab/Fakenewsdata.csv')
data.head()
```

```python
X = data.drop('label',axis = 1)
X.head()
```

```python
data = data.dropna()
```

```python
data.reset_index(inplace = True)  ## Since we had droped null values we are reseting indexes of data frame.
```

```python
X1 = data.drop('label' , axis = 1)
X1.head()
```

```python
Y1 = data['label']
Y1.head()
```

```python
X1.shape
```

```python
Y1.shape
```

```python
voc_size = 5000
```

```python
import nltk
```

```python
import re
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
PS = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(X1)):
  review = re.sub('[^a-zA-Z]'," ",X1['title'][i])
  review = review.lower()
  review = review.split()

  review = [PS.stem(word) for word in review if not word in stopwords.words('english')]
  review = " ".join(review)
  corpus.append(review)
```

```python
corpus
```

```python
onehot_rep = [one_hot(words,voc_size)for words in corpus]
onehot_rep
```

```python
sent_length = 20
embedded_docs = pad_sequences(onehot_rep,padding = 'pre', maxlen = sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
embedding_vector_features = 40
model = Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length = sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation = 'sigmoid'))
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
print(model.summary())
```

```python
len(embedded_docs),Y1.shape
```

```python
X_final = np.array(embedded_docs)
Y_final = np.array(Y1) 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_final,Y_final, test_size = 0.33, random_state = 42)
```

```python
model.fit(X_train,Y_train, batch_size = 64, epochs = 10, validation_data=(X_test,Y_test))
```

```python
Y_pred = model.predict_classes(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test,Y_pred)
```

```python
confusion_matrix(Y_pred,Y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred)
```

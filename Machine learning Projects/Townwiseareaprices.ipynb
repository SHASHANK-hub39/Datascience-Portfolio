```python
import xlrd
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_excel("Townwisearearates.xlsx")
df
```

```python
pd.get_dummies(df.Town)
```

```python
dummies = pd.get_dummies(df.Town)
DF = pd.concat([df,dummies],axis="columns")
DF
```

```python
FinalDF = DF.drop(['Town','Palvancha'],axis='columns')
FinalDF
```

```python
X = FinalDF.drop(['Price'],axis='columns')
X
```

```python
Y = FinalDF.Price
Y
```

```python
reg = linear_model.LinearRegression()
reg.fit(X,Y)
```

```python
reg.predict([[2800,0,1]])
```

```python
reg.predict([[3400,0,0]])
```

```python
reg.score(X,Y)
```
